### Grad-CAM et dÃ©couverte des zones sÃ©mantiques (tf-explain)

Dans ce TP, nous allons :
- comprendre comment un CNN (rÃ©seau de neurones convolutionnel) â€œvoitâ€ une image ;
- utiliser la bibliothÃ¨que **tf-explain** pour gÃ©nÃ©rer une carte Grad-CAM ;
- comparer la prÃ©diction du modÃ¨le avec la zone rÃ©ellement utilisÃ©e ;
- dÃ©couvrir le biais classique du â€œloup dans la neigeâ€.

**But :** comprendre que lâ€™IA peut parfois avoir raisonâ€¦ pour de mauvaises raisons.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import resnet50
from tensorflow.keras.applications.resnet50 import decode_predictions, preprocess_input
from tensorflow.keras.preprocessing import image

# pip install tf-explain si nÃ©cessaire
from tf_explain.core.grad_cam import GradCAM
```

```python
# TODO : ajouter une image de loup dans la neige
img_path = "loup_neige.jpg"  # Ã  adapter

img = image.load_img(img_path, target_size=(224,224))
img_array = image.img_to_array(img)
img_input = np.expand_dims(img_array.copy(), axis=0)
img_preprocessed = preprocess_input(img_input)

plt.imshow(img_array.astype("uint8"))
plt.axis("off")
plt.title("Image originale (loup/neige)")
```

### â“ Questions 1â€“2
1. Que pense lâ€™IA quâ€™il y a sur cette image ?  
2. Sur quelles zones, selon toi, lâ€™IA devrait-elle se concentrer ?

```python
model = resnet50.ResNet50(weights="imagenet")
preds = model.predict(img_preprocessed)
decode_predictions(preds, top=3)[0]
```

### â“ Questions 3â€“4
3. Le modÃ¨le a-t-il raison dans sa prÃ©diction ?  
4. Ã€ ton avis, comment un rÃ©seau fait-il pour distinguer un loup dâ€™un chien ?

```python
data = (np.array([img_array.astype("uint8")]), None)

explainer = GradCAM()
layer_name = "conv5_block3_out"

grid = explainer.explain(
    validation_data=data,
    model=model,
    class_index=int(np.argmax(preds)),
    layer_name=layer_name
)

plt.figure(figsize=(10,4))
plt.imshow(grid)
plt.axis("off")
plt.title("Grad-CAM : zones importantes")
```

### â“ Questions 5â€“7
5. Quelles zones sont rouges (importance Ã©levÃ©e) ?  
6. Lâ€™IA â€œregardeâ€-t-elle rÃ©ellement lâ€™animal ?  
7. Observe-tu un indice visuel qui pourrait tromper lâ€™IA ?

```python
# TODO : ajouter une image de chien dans la neige
img2_path = "chien_neige.jpg"  # Ã  adapter

img2 = image.load_img(img2_path, target_size=(224,224))
img2_array = image.img_to_array(img2)
img2_input = np.expand_dims(img2_array.copy(), axis=0)
img2_preprocessed = preprocess_input(img2_input)

preds2 = model.predict(img2_preprocessed)
decode_predictions(preds2, top=3)[0]
```

```python
data2 = (np.array([img2_array.astype("uint8")]), None)
grid2 = explainer.explain(
    validation_data=data2,
    model=model,
    class_index=int(np.argmax(preds2)),
    layer_name=layer_name
)

plt.figure(figsize=(10,4))
plt.imshow(grid2)
plt.axis("off")
plt.title("Grad-CAM : image chien/neige")
```
