## 5. Feature Maps ‚Äì Comprendre ce qu‚Äôil se passe dans le r√©seau
Les *feature maps* montrent ce que les couches internes retiennent de l‚Äôimage.

```python
layer_idx = 10
target_layer = model.model.model[layer_idx]

feature_maps={}
def capture(module,inp,out):
    feature_maps["fm"]=out.detach().cpu()

h = target_layer.register_forward_hook(capture)
_ = model(img_rgb)
h.remove()

fm = feature_maps["fm"][0]
C,H,W = fm.shape
print("Taille :", fm.shape)

fig,axes=plt.subplots(1,6,figsize=(15,3))
for i in range(6):
    ch = int(i*C/6)
    fmap=fm[ch]
    fmap=(fmap-fmap.min())/(fmap.max()-fmap.min()+1e-6)
    axes[i].imshow(fmap,cmap="inferno")
    axes[i].axis("off")
plt.show()
```
